```python
import sys
sys.path.insert(0, '/home/yw440/MXGPY/mxgpy')
sys.path.insert(0, '/home/yw440/MyRGPs/')
import os
os.environ['MXNET_ENGINE_TYPE'] = 'NaiveEngine'

import scipy.io
from pylab import *
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import GPy
import numpy as np
np.random.seed(101)

from mxgpy.rgp.rgp import RGP 
from mxgpy.kernel.rbf import RBF
import mxnet as mx
ctx = mx.gpu()
mx.Context.default_ctx = ctx

from mxgpy.util import positive_transform_reverse_numpy
import helper
```

```python
walk_train_motions = ['01', '02', '03', '04']
run_train_motions = ['17', '18', '19', '20']
walk_test_motions = ['05', '06', '07', '08']
run_test_motions = ['21', '22', '23', '24']

subject = '35'
y_bones = ['root', 'lfemur', 'ltibia', 'lfoot', 'ltoes', 'rfemur', 'rtibia', 'rfoot', 'rtoes', 
             'lowerback', 'upperback', 'thorax', 'lowerneck', 'upperneck', 'head', 
             'lclavicle', 'lhumerus', 'lradius', 'lwrist', 'lhand', 'lfingers', 'lthumb',
            'rclavicle', 'rhumerus', 'rradius', 'rwrist', 'rhand', 'rfingers', 'rthumb']
# list of control signal in order
u_bones = ['None', 'root', 'lfemur', 'ltibia', 'lfoot', 'root', 'rfemur', 'rtibia', 'rfoot', 
          'root', 'lowerback', 'upperback', 'thorax', 'lowerneck', 'upperneck',
          'thorax', 'lclavicle', 'lhumerus', 'lradius', 'lwrist', 'lhand', 'lwrist',
          'thorax', 'rclavicle', 'rhumerus', 'rradius', 'rwrist', 'rhand', 'rwrist' ]

m1 = GPy.util.datasets.cmu_mocap(subject, walk_train_motions, walk_test_motions, sample_every=4, data_set='cmu_mocap')
m2 = GPy.util.datasets.cmu_mocap(subject, run_train_motions, run_test_motions, sample_every=2, data_set='cmu_mocap')
## Use different sample rate so the training sequence length of walking and running stay similar

Y  = np.vstack([m1['Y'], m2['Y']])
fill1 = np.zeros((np.shape(m1['lbls'])[0],4))
fill2 = np.zeros((np.shape(m2['lbls'])[0],4))
labels = np.vstack( [np.hstack([m1['lbls'], fill1]), np.hstack([fill2, m2['lbls']])] )

y_train_all = helper.extract_independent_bones(Y, labels) # list of lists of arrays
y_test_walk_joint = helper.extract_independent_bones(m1['Ytest'], m1['lblstest'])
y_test_run_joint = helper.extract_independent_bones(m2['Ytest'], m2['lblstest'])


## Reshape the sequences   
y_train_list = helper.reshape(y_train_all)
_, y_train_mean, y_train_std = helper.normalise_data([Y])
y_train_mean = np.asarray(y_train_mean)
y_train_std = np.asarray(y_train_std)
y_train = []
for item in y_train_list:
    y_train.append( (item-y_train_mean)/(y_train_std+1e-5) )

y_test_walk_list = helper.reshape(y_test_walk_joint)
y_test_run_list = helper.reshape(y_test_run_joint)
y_test_walk, y_test_run =[],[]

_, y_test_mean, y_test_std = helper.normalise_data([np.vstack((m1['Ytest'],m2['Ytest']))])
y_test_mean = np.asarray(y_test_mean)
y_test_std = np.asarray(y_test_std)

for item in y_test_walk_list:
    y_test_walk.append( (item-y_test_mean)/(y_test_std+1e-5) )
    
for item in y_test_run_list:
    y_test_run.append( (item-y_test_mean)/(y_test_std+1e-5) )
 
```

```python
np.shape(y_train_all)
```

```python

mean_list, std_list = [], []
for item in y_train_all:
    out, mean, std = helper.normalise_data(item)
    mean_list.append(mean)
    std_list.append(std)
    
## Reshape the sequences
mean = helper.reshape(mean_list)
std = helper.reshape(std_list)

train_mean = [x.mean() for x in mean];
train_std = [x.mean() for x in std]

u_train = []
for i, item in enumerate(y_train):
    control = np.ones((np.shape(item)[0], 2))
    control[:,0] = control[:,0]*train_mean[i]
    control[:,0] = control[:,0]*train_std[i]
    u_train.append(control)
    
```

```python
np.shape(u_train)#, train_std)
```

```python
walk_mean_list, walk_std_list = [], []
for item in y_test_walk_joint:
    out, mean, std = helper.normalise_data(item)
    walk_mean_list.append(mean)
    walk_std_list.append(std)
    
## Reshape the sequences
walk_mean = helper.reshape(walk_mean_list)
walk_std = helper.reshape(walk_std_list)



run_mean_list, run_std_list = [], []
for item in y_test_run_joint:
    out, mean, std = helper.normalise_data(item)
    run_mean_list.append(mean)
    run_std_list.append(std)
    
## Reshape the sequences
run_mean = helper.reshape(run_mean_list)
run_std = helper.reshape(run_std_list)


test_walk_mean = [x.mean() for x in walk_mean]; print('walk mean: ', test_walk_mean)
test_walk_std = [x.mean() for x in walk_std]; print('walk_std: ', test_walk_std)

test_run_mean = [x.mean() for x in run_mean]; print('run mean: ', test_run_mean)
test_run_std = [x.mean() for x in run_std]; print('run std: ', test_run_std)




u_test_walk = []
for i, item in enumerate(y_test_walk):
    control = np.ones((np.shape(item)[0],2))
    control[:,0] = control[:,0]*test_walk_mean[i]
    control[:,1] = control[:,1]*test_walk_std[i]
    u_test_walk.append(control)
    
u_test_run = []
for i, item in enumerate(y_test_run):
    control = np.ones((np.shape(item)[0],2))
    control[:,0] = control[:,0]*test_run_mean[i]
    control[:,1] = control[:,1]*test_run_std[i]
    u_test_run.append(control)
```

```python
## --- Training --- ##

YD = np.shape(y_train[0])[-1]
UD = np.shape(u_train[0])[-1]
HD = 1
win_in = 20
win_hidden = 20
win_out = 20

m = RGP(wins=[win_hidden, win_out], with_control=True, X_dims=[HD, YD], U_dim=UD, U_win=win_in, num_inducing=100,
        kernels=[RBF(win_hidden*HD+win_in*UD, ARD=True, dtype=np.float64, name='rbf0'), 
                 RBF((win_hidden+1)*HD+win_out*YD, ARD=True, dtype=np.float64, name='rbf1')],
        nSamples=5, q_hidden_units=[200,300,200], h0_hidden_units=[200,300,200], ctx=ctx, dtype=np.float64)
    
## Initialize parameters
m.collect_params()['rbf0_lengthscale'].set_data(positive_transform_reverse_numpy(np.ones((win_hidden*HD+win_in*UD,))*3,np.float64))
m.collect_params()['rbf1_lengthscale'].set_data(positive_transform_reverse_numpy(np.ones(((win_hidden+1)*HD+win_out*YD,))*10,np.float64))
m.collect_params()['layer_0_noise_var'].set_data(positive_transform_reverse_numpy(np.array([1e-4]),np.float64))
m.collect_params()['layer_1_noise_var'].set_data(positive_transform_reverse_numpy(np.array([np.var(np.vstack(y_train))*0.001]),np.float64))    
  
## Train model
m.optimize(Y=y_train, U=u_train, max_iters=200, step_rate=0.025)
m.optimize(Y=y_train, U=u_train, max_iters=300, step_rate=0.02)
m.optimize(Y=y_train, U=u_train, max_iters=1500, step_rate=0.01)
## Save model
m.save_params('./models/last_mixed')
m.optimize(Y=y_train, U=u_train, max_iters=1000, step_rate=0.005)
    
# ## Save model
# m.save_params('./models/control_mixed_normalise3d') 
```

```python
def test_run(model, seq, win_out,  test_data, HD, win_in=None,control_signal=None, MEAN_PRED=True, with_control=True):
    if with_control:
        YD = np.shape(test_data[0])[-1]; print('Observationa dimension:', YD)
        UD = np.shape(control_signal[seq])[-1] ; print('Control signal dimension:', UD)
        
        m = RGP(wins=[win_out, 0], with_control=True, X_dims=[HD, YD], U_dim=UD, U_win=win_in, num_inducing=100,
                kernels=[RBF(win_out*HD+win_in*UD, ARD=True, dtype=np.float64, name='rbf0'), 
                         RBF((win_out+1)*HD, ARD=True, dtype=np.float64, name='rbf1')],
                nSamples=5, q_hidden_units=[200,200], h0_hidden_units=[200,200], ctx=ctx, dtype=np.float64)
        
        m.load_params('./models/'+model)
    
        y_pd = m.freerun(test_data[seq].shape[0]-win_out, init_X=test_data[seq][:win_out,:],
                         U=control_signal[seq], mean_predict=False, nSamples=100)
    
        y_pd_mean = y_pd.mean(0) # the first dimention of the matrix is number of samples
        y_pd_std = y_pd.std(0)
    else:
        YD = np.shape(test_data[0])[-1]; print('Observationa dimension:', YD)
        
    
        m = RGP(wins=[win_out, win_out], with_control=False, X_dims=[HD, YD], num_inducing=100,
                kernels=[RBF(win_out*HD, ARD=True, dtype=np.float64, name='rbf0'), 
                         RBF((win_out+1)*HD+win_out*YD, ARD=True,dtype=np.float64, name='rbf1')],
                nSamples=5, q_hidden_units=[200, 200], h0_hidden_units=[200,200], ctx=ctx, dtype=np.float64)
        m.load_params('./models/'+model)
    
        y_pd = m.freerun(test_data[seq].shape[0]-win_out, init_X=test_data[seq][:win_out,:], mean_predict=False, nSamples=100)
    
        y_pd_mean = y_pd.mean(0) # the first dimention of the matrix is number of samples
        y_pd_std = y_pd.std(0)
    
    if MEAN_PRED:
        return y_pd_mean
    else:
        return y_pd_mean + y_pd_std*np.random.normal(0,0.5,size=(y_pd_mean.shape[0], y_pd_mean.shape[1]))
```

```python
def compute_MSE(y_test, pds):
    MSE = []
    for i, pd in enumerate(pds):
        tmp = pd - y_test[i]
        MSE.append( tmp.dot(tmp.T).diagonal().mean()/62 )
    return MSE

## Global mean & variance
print(compute_MSE(y_test_walk, y_pds))
```

```python
def test_run(model, seq, win_out, test_data, HD, win_in=None,control_signal=None, MEAN_PRED=True, with_control=True):
    if with_control:
        YD = np.shape(test_data[0])[-1]; print('Observationa dimension:', YD)
        UD = np.shape(control_signal[0])[-1] ; print('Control signal dimension:', UD)
        length = np.shape(control_signal[0])[0]; print('sequence length: ', length )
        
        m = RGP(wins=[win_out, 0], with_control=True, X_dims=[HD, YD], U_dim=UD, U_win=win_in, num_inducing=100,
                kernels=[RBF(win_out*HD+win_in*UD, ARD=True, dtype=np.float64, name='rbf0'), 
                         RBF((win_out+1)*HD, ARD=True, dtype=np.float64, name='rbf1')],
                nSamples=5, q_hidden_units=[200,200], h0_hidden_units=[200,200], ctx=ctx, dtype=np.float64)
        
        m.load_params('./models/'+model)
    
        y_pd = m.freerun(test_data[seq].shape[0]-win_out, init_X=test_data[seq][:win_out,:],U=control_signal[seq], mean_predict=False, nSamples=100)
#         y_sin = m.freerun(1000, init_X=yx[:win_out], mean_predict=False, nSamples=100)

        y_pd_mean = y_pd.mean(0) # the first dimention of the matrix is number of samples
        y_pd_std = y_pd.std(0)

    if MEAN_PRED:
        return y_pd_mean
    else:
        return y_pd_mean + y_pd_std*np.random.normal(0,0.5,size=(y_pd_mean.shape[0], y_pd_mean.shape[1]))
```
